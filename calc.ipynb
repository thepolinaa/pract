```python
# 1
def calculate_string_expression(expression):
    numbers = {
        'ноль': 0, 'один': 1, 'два': 2, 'три': 3, 'четыре': 4, 'пять': 5,
        'шесть': 6, 'семь': 7, 'восемь': 8, 'девять': 9, 'десять': 10,
        'одиннадцать': 11, 'двенадцать': 12, 'тринадцать': 13, 'четырнадцать': 14,
        'пятнадцать': 15, 'шестнадцать': 16, 'семнадцать': 17, 'восемнадцать': 18,
        'девятнадцать': 19, 'двадцать': 20, 'тридцать': 30, 'сорок': 40,
        'пятьдесят': 50, 'шестьдесят': 60, 'семьдесят': 70, 'восемьдесят': 80, 'девяносто': 90 }
    operations = {
        'плюс': lambda x, y: x + y,
        'минус': lambda x, y: x - y,
        'умножить': lambda x, y: x * y }
    operand_1, operation, operand_2 = expression.split()
    result = operations[operation](numbers[operand_1], numbers[operand_2])
    if result in numbers.values():
        for word, number in numbers.items():
            if number == result:
                return word
    else:
        return 'результат не попадает в диапазон от 0 до 99'
print(calculate_string_expression('двадцать пять плюс тринадцать'))  # возвращает "тридцать восемь"
```

```python
# 2
num_dict = {
    "ноль": 0, "один": 1, "два": 2, "три": 3, "четыре": 4, "пять": 5, "шесть": 6, "семь": 7, "восемь": 8, "девять": 9,
    "десять": 10, "одиннадцать": 11, "двенадцать": 12, "тринадцать": 13, "четырнадцать": 14, "пятнадцать": 15,
    "шестнадцать": 16, "семнадцать": 17, "восемнадцать": 18, "девятнадцать": 19, "двадцать": 20, "тридцать": 30,
    "сорок": 40, "пятьдесят": 50, "шестьдесят": 60, "семьдесят": 70, "восемьдесят": 80, "девяносто": 90 }
operation_dict = {
    "плюс": "+",
    "минус": "-",
    "умножить": "*",
    "разделить": "/",
    "остаток от деления": "%" }
def calc(expression):
    parts = expression.split()
    if len(parts) < 3:
        return "неверный формат выражения"
    num1 = 0
    num2 = 0
    op = ""
    for word in parts:
        if word in num_dict:
            if op == "":
                num1 = num1 * 10 + num_dict[word]
            else:
                num2 = num2 * 10 + num_dict[word]
        elif word in operation_dict:
            op = operation_dict[word]
        else:
            return "неверный формат выражения"
    try:
        if op == "+":
            result = num1 + num2
        elif op == "-":
            result = num1 - num2
        elif op == "*":
            result = num1 * num2
        elif op == "/":
            result = num1 / num2
        elif op == "%":
            result = num1 % num2
    except ZeroDivisionError:
        return "деление на ноль"
    if "." in str(result):
        result_int, result_dec = str(result).split(".")
        result_int = int(result_int)
        if result_dec[2] >= "5":
            result_int += 1
    else:
        result_int = result
        result_dec = 0
    result_text = ""
    for key, value in num_dict.items():
        if value == result_int:
            result_text = key
            break
    if result_dec != 0:
        result_text += " и "
        for key, value in num_dict.items():
            if value * 0.01 == int(result_dec):
                result_text += key
                break
        result_text += " сотых"
    return result_text
print(calc("двадцать пять плюс тринадцать"))  # "тридцать восемь"
print(calc("сорок один и тридцать одна сотая разделить на семнадцать"))  # "два и сорок три сотых" 
```

```python
# 3
from num2words import num2words
def calc(expression):
    words = expression.split()
    num1 = words[0] + " " + words[1] if words[1] == "и" else words[0]
    num2 = words[-2] + " " + words[-1] if words[-2] == "и" else words[-1]
    operation = words[len(words)//2]
    if "." in num1:
        num1_integer, num1_decimal = num1.split("и")
        if num1_integer.isdigit():
            num1_integer = int(num1_integer)
        else:
            num1_integer = int(num2words(num1_integer, lang='ru'))
        num1_decimal = int(num1_decimal)/100
        num1 = num1_integer + num1_decimal
    elif num1.isdigit():
        num1 = int(num1)
    else:
        num1 = num2words(num1, lang='ru')
    if "." in num2:
        num2_integer, num2_decimal = num2.split("и")
        if num2_integer.isdigit():
            num2_integer = int(num2_integer)
        else:
            num2_integer = int(num2words(num2_integer, lang='ru'))
        num2_decimal = int(num2_decimal)/100
        num2 = num2_integer + num2_decimal
    elif num2.isdigit():
        num2 = int(num2)
    else:
        num2 = num2words(num2, lang='ru')
    if operation == "плюс":
        result = num1 + num2
    elif operation == "минус":
        result = num1 - num2
    elif operation == "умножить":
        result = num1 * num2
    elif operation == "разделить":
        result = num1 / num2
        result_str = str(result)
        decimal_part, fraction_part = result_str.split(".")
        if len(fraction_part) > 2:
            period = find_decimal_period(fraction_part)
            return str(decimal_part) + " и " + fraction_part[:2] + " в периоде " + period
    return num2words(result, lang='ru')
def find_decimal_period(fraction):
    period = ""
    for i in range(1, 5):
        for j in range(i, len(fraction), i):
            if fraction[j: j+i] == fraction[:i]:
                period = fraction[:i]
                break
        if period:
            break
    return period
print(calc("двадцать пять плюс тринадцать"))  # "тридцать восемь"
print(calc("девятнадцать и восемьдесят две сотых разделить на девяносто девять"))  # "ноль и двадцать сотых и ноль два в периоде"
```

```python
# 4
import sympy
from num2words import num2words
def calculate(expression):
    operations = {
        "плюс": '+',
        "минус": '-',
        "умножить": '*',
        "на": '*' }
    for op, sym in operations.items():
        expression = expression.replace(op, sym)
    result = sympy.sympify(expression)
    return num2words(result)
print(calculate("пять плюс два умножить на три минус один"))
```

```python
# 5
from num2words import num2words
import sympy
def calculate(expression):
    operations = {
        "плюс": '+',
        "минус": '-',
        "умножить": '*',
        "скобка открывается": '(',
        "скобка закрывается": ')',
        "на": '*' }
    for op, sym in operations.items():
        expression = expression.replace(op, sym)
    result = sympy.sympify(expression)
    return num2words(result)
print(calculate("скобка открывается пять плюс два скобка закрывается умножить на три минус один"))
```

```python
# 7
from fractions import Fraction
def calc(expression):
    numbers = { "ноль": 0, "один": 1, "два": 2, "три": 3, "четыре": 4, "пять": 5, "шесть": 6, "семь": 7,
               "восемь": 8, "девять": 9, "десять": 10, "одиннадцать": 11, "двенадцать": 12, "тринадцать": 13,
               "четырнадцать": 14, "пятнадцать": 15, "шестнадцать": 16, "семнадцать": 17, "восемнадцать": 18,
               "девятнадцать": 19, "двадцать": 20, "тридцать": 30, "сорок": 40, "пятьдесят": 50, "шестьдесят": 60,
               "семьдесят": 70, "восемьдесят": 80, "девяносто": 90 }
    split_expr = expression.split()
    fractions = []
    current_fraction = []
    for word in split_expr:
        if word in numbers:
            current_fraction.append(numbers[word])
        elif word == "один":
            current_fraction.append(1)
        elif word == "и":
            fractions.append(Fraction(current_fraction[0], current_fraction[1]))
            current_fraction = []
    operation = split_expr[-2]
    if operation == "плюс":
        result_fraction = sum(fractions)
    elif operation == "минус":
        result_fraction = fractions[0] - fractions[1]
    elif operation == "умножить":
        result_fraction = fractions[0] * fractions[1]
    if result_fraction.numerator < result_fraction.denominator:
        return str(result_fraction)
    else:
        whole_part = result_fraction.numerator // result_fraction.denominator
        remainder = result_fraction - whole_part
        if whole_part > 0:
            return f"{whole_part} и {remainder}"
        else:
            return str(remainder)
print(calc("один и четыре пятых плюс шесть седьмых"))  # "два и двадцать три тридцать пятых"
```
